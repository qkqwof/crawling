## 네이버쇼핑 크롤링

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://search.shopping.naver.com/best100v2/main.nhn")
soup = BeautifulSoup(res.content,"html.parser")
data = soup.select("#popular_srch_lst > li a")
for i in data:
    print(i.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://search.shopping.naver.com/best100v2/main.nhn")
soup = BeautifulSoup(res.content,"html.parser")
mydata = soup.select("#popular_srch_lst > li > span.txt")
for i in mydata:
    print(i.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

site_list = ["https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000",
             "https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000002"]

for site in site_list:
    res = requests.get("https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000")
    soup = BeautifulSoup(res.content,"html.parser")
    data = soup.select("#productListArea > ul > li > p > a")
    print(site)
    for i in data:
        print(i.get_text())
```

## 네이버 주식

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://finance.naver.com/sise/")
soup = BeautifulSoup(res.content,"html.parser")
data = soup.select("#popularItemList > li > a")
for i in data:
    print(i.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://finance.naver.com/sise/")
soup = BeautifulSoup(res.content,"html.parser")
data = soup.select("#popularItemList>li>a")
for i in data:
    print(i.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://finance.naver.com/sise/")
soup = BeautifulSoup(res.content,"html.parser")
data = soup.select("div.rgt > ul.lst_major > li")

for i,d in enumerate(data):
    print(str(i+1)+".",d.find("a").get_text().strip(), d.find("span").get_text(),d.find("em").get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://finance.naver.com/sise/")
soup = BeautifulSoup(res.content,"html.parser")
data = soup.select("div.rgt > ul.lst_major > li")
for i in data:
    print(i.find("a").get_text(),i.find("span").get_text(), i.find("em").get_text())
```
